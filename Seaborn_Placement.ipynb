# Bivariate Analysis : Distribution plot 

```python
# joint plot
```

```python
sns.jointplot(x='etest_p',y='mba_p',data=dataset)
```

```python
sns.jointplot(x='mba_p',y='salary',data=dataset,kind='kde')
```

```python
sns.scatterplot(x='hsc_p',y='etest_p',data=dataset)
```

```python
sns.jointplot(x='etest_p',y='salary',data=dataset,kind='hex')
```

```python
# pairplot
```

```python
sns.pairplot(dataset,hue='gender',diag_kind='kde',kind='scatter',palette='bright')
```

# For categorical columns 


```python
# stirpplot
```

```python
sns.stripplot(x='degree_t',y='degree_p',data=dataset)
```

```python
sns.stripplot(x='degree_t',y='degree_p',data=dataset,jitter=True)
```

```python
# swarmplot - used same as stripplot-useful in scenarios where you want to avoid overlapping points
# and provide a better representation of the distribution.
```

```python
sns.swarmplot(x='degree_t',y='degree_p',data=dataset)
```

# Distribution observation

```python
# violin plot
```

```python
sns.violinplot(x='specialisation',y='mba_p',data=dataset,palette='bright')
```

```python
sns.violinplot(x='specialisation',y='mba_p',data=dataset,palette='bright',hue='gender')
```

```python
# box plot - used to display percentile concept,IQR,quartile,outliers
```

```python
sns.boxplot(x='degree_t',y='degree_p',data=dataset,hue='gender')
```

```python
# bar plot
```

```python
sns.barplot(x='degree_t',y='degree_p',data=dataset,hue='gender')
```

```python
# countplot - univariate analysis - used to find the count
```

```python
sns.countplot(x='degree_t',data=dataset,hue='gender')
```

```python
# pointplot=  Point plots are particularly useful when you want to 
#compare the mean or median of a numeric variable across different categories.
```

```python
sns.pointplot(x='degree_t',y='degree_p',data=dataset,hue='gender')
```

```python
sns.catplot(x='degree_t',y='degree_p',hue='gender',data=dataset,kind='violin')
```
